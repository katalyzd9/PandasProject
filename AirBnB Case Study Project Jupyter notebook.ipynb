```python
# Which hosts are the busiest and based on their reviews?

busyhost = air_bnb.groupby('host_id')
host = busyhost[['number_of_reviews']].count()
sorted_hosts = host.sort_values(['number_of_reviews'], ascending=False)
sorted_hosts
```

```python
#Which neighorhood group has the highest average price?
price = air_bnb.groupby('neighbourhood_group')['price'].mean()
price
```
